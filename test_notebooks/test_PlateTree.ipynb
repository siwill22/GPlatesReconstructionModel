## PlateTree class usage examples

```python
%load_ext autoreload
%autoreload 2

from gprm import PlateTree
from gprm.datasets import Reconstructions
```

```python
M2019 = Reconstructions.fetch_Muller2019()
```

```python
tree_object = PlateTree(M2019)

tree_object.plot_snapshot(200.)
```

```python
tree_object.plot_snapshot(50., figsize=(20,10))
```

```python
import pygmt

reconstruction_time = 250.

region = 'd'
projection="N6i"

fig = pygmt.Figure()
fig.basemap(region=region, projection=projection, frame='afg')

M2019.polygon_snapshot('continents', reconstruction_time).plot(fig, transparency=30)

tree_object.plot_gmt(fig, reconstruction_time)

fig.savefig('../plots/platetree.png')
fig.show(width=800)
```
